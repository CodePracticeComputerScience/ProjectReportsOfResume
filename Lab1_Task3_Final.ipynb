## Run the Decision Tree classifier on Image Segmentation data, Large Soybean Database, and Thyroid disease records with a 72% and 28% split of training and test data respectively. 

### Image Segmentation data
#### EDA 
#### Attributes 19
#### Classes 7
#### Data floats,integers

```python
from scipy.io import arff
import pandas as pd

data = arff.loadarff('data/segment.arff')
df = pd.DataFrame(data[0])

df.head()
```

```python
from datacleaner import autoclean

my_clean_data = autoclean(df)

print(my_clean_data)
```

```python
x = df[df.columns.drop('class')]
print(x)
```

```python
y = df['class']
print(y)
```

```python
# import Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

clf = tree.DecisionTreeClassifier(random_state=0)#random_state will give consistent results
clf = clf.fit(x,y)
```

```python
#number of nodes
treeObj = clf.tree_
print treeObj.node_count
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
```

```python
Y_pred = cross_val_predict(clf,x,y,cv=10)
print(Y_pred)
conf_mat = confusion_matrix(y,Y_pred)
print (conf_mat)
```

```python
#accuracy
from sklearn import metrics
print metrics.accuracy_score(y,Y_pred)
```

```python
#error rate

e = metrics.accuracy_score(y,Y_pred)
print 1-e
```
