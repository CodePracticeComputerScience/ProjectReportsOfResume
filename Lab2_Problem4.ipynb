```python
#load plants dataset
import numpy as np
import csv
with open('PlantsDF.csv', 'r') as f:
    plants = list(csv.reader(f, delimiter=','))
plants
```

This dataset is having multivariate values and dataset is not in proper format ,so first have to create dataframe out of it.State abbreviations frequency can be replaced with binary token.



```python
#first I am getting all distinct state abbreviations from the csv file.

#state abbreviations are my column headings and latin names are row headings.

#in each latin name when I have present value for state abbr. I am making that as 1 and other than that evrything else is 0.

col_labels_set=set([])

for item in range(len(plants)):
    #print(item)
    for inner_list in range(len(plants[item])):
        if inner_list == 0:
            continue
        col_labels_set.add(plants[item][inner_list])
col_labels=list(col_labels_set)


row_labels=[]
for item in range(len(plants)):
    row_labels.append(plants[item][0])
#print(row_labels)


import pandas as pd

df=pd.DataFrame(0,index = row_labels,columns=  sorted(col_labels))

```

```python

# Iterating each line from csv file and if column heading matches with state abbr. put 1 there.


with open('Plants.csv','r') as f :
    for cnt, line in enumerate(f):
        line=line.replace("\n","")
        splited_line=line.split(",")
        row_index=splited_line[0]
        #print(">>>>>>>>>>>>>>>",row_index)
        splited_line.remove(row_index)
        #print(splited_line)
        #print(item)
        for item in splited_line:
         df.loc[row_index][item]=1
         #print(df.loc[row_index][item])
        

df.to_csv('PlantsDF.csv')
```

### AffinityPropagation

```python
#Memory error


from sklearn.cluster import AffinityPropagation
af = AffinityPropagation(damping=0.5, max_iter=200, convergence_iter=15, copy=True, preference=None, verbose=False).fit(X)

cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_

n_clusters_ = len(cluster_centers_indices)

print('Estimated number of clusters: %d' % n_clusters_)
```
