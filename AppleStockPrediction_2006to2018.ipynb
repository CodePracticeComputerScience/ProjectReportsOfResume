### Simple model. 

```python
#Reference:https://keras.io/getting-started/sequential-model-guide/
#Reference:https://machinelearningmastery.com/keras-functional-api-deep-learning/
#https://www.youtube.com/watch?v=Msk-A99oo10



from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
#"add method" adds layers and in this simple method we have 
#1 input layer 
# 1 internal layer 
# and output layer.

#return_sequences = False will not interlink input layers.

def simple_model():
    model = Sequential()

# units = 10 means LSTM has 10 RNN(Recurrent neural network blocks)
#This RNN accepts shape 1 as input that is why converting X_train in shape 1 array.

    model.add(LSTM(units = 10, return_sequences = False, input_shape = (X_train.shape[1], 1)))
    
#Reference:https://keras.io/layers/core/
#Units = 1 indicates dimensionality of the output which is only one here.
#all other parameters are default

    model.add(Dense(units = 1))
    return model
```

## Solution : Complex Model

Using  a stacked LSTM(i.e multiple LSTM layers instead of 1). 
and also increase the number of units per LSTM cell to 50.

```python
# We are adding 4 layers and changing units from 10 to 50 in LSTM.
```

```python
def complex_model():
    model = Sequential()

    # Adding the first LSTM layer
    model.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))

    # Adding a second LSTM layer
    model.add(LSTM(units = 50, return_sequences = True))
    
    # Adding a third LSTM layer
    model.add(LSTM(units = 50, return_sequences = True))

    # Adding a fourth LSTM layer
    model.add(LSTM(units = 50))

    # Adding the output layer
    model.add(Dense(units = 1))
    return model
```

```python
complexmodel = complex_model()
```

```python
complexmodel.summary()
```

```python
history = compile_and_run(complexmodel, epochs=20)
```

```python
plot_metrics(history)
```

As expected, we see that the overall loss of the model reducing until a minima is reached. After that, the loss just stays constant.
